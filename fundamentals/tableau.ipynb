## 3. Tableau Worksheet Basics

**Views -** A view is a visualization (or viz) that you create in Tableau. A viz might be a chart, a graph, a map, a plot, or even a text table. Every view that you build in Tableau should start with a question. What do you want to know?

**Dashboards -** A dashboard is a collection of several views, letting you compare a variety of data simultaneously.

**Stories -** In Tableau, a story is a sequence of visualizations that work together to convey information. You can create stories to tell a data narrative, provide context, demonstrate how decisions relate to outcomes, or to simply make a compelling case.

[Stories in Detail](https://help.tableau.com/current/pro/desktop/en-us/stories.htm)

**Aggregation -** Row-level data rolled up to a higher category, such as sum of sales or total profit. Tableau does this automatically so you can break data down to the level of detail that you want work with.

### [Dimensions and Measures in Detail](https://help.tableau.com/current/pro/desktop/en-us/datafields_typesandroles.htm)

**Measures -** Quantiative (continuos) numerical data denoted by a green # sign next to the field name (generally the column header). These data points are dependent variables.

### Tableau Order of Operations
Tableau has a specific order of operations that it follows appropriately called the 'Tableau Order of Operations' (also known as the query pipeline). This is the order that Tableau perfoms actions behind the the scenes (such as applying filters to a view). 

In some cases (like filters) a newly applied action may not show up on because it has been inserted outside of the regular order of operations. To fix issues like these, Tableau has to be told the order (context) you want to perform them in. 

So if you have say a Context Filter and a Top N Filter on your workbook, the top N Filter will not show up unless context is added to move it up in the order of operations

## 7b. [Level of Detail (LOD)](https://help.tableau.com/current/pro/desktop/en-us/calculations_calculatedfields_lod.htm)

[Decent Vid Tutorial](https://www.youtube.com/watch?v=yush1yNz9VM&ab_channel=sqlbelle)

[Really Clear Tutorial](https://vizartpandey.com/level-of-detail-expressions-in-tableau/)

[LOD Tutorial](https://interworks.com/blog/rcurtis/2016/03/21/tableau-deep-dive-lod-introduction-detail/)

Level of Detail is a syntax in Tableau that simplifies and extends Tableaus calculations by making it possible to address detailed queries directly. This allows for a single visualization to answer questions involving multiple levels of granularity. As a general rule-of-thumb, in Tableau, the more granularity the less aggregation there is and vice-versa. 

For a workseet in Tableau, there are 3 areas that primarily determine the detail of the view: 

**1. Row & Column Shelves**

**Detail** is defined by the dimensions you use to segment your chosen measures. For example, if you add a single measure (sales in the example) and a single dimension (state in the example) to the shelves like so:

<img src="data/images/tab14.png" width="300">

Then you are getting a single level of detail which will aggregate the sum of sales by the State. 

If another dimension is added (Category in the example), then the level of detail increases and there is more ganularity and less aggregation on the results, see diagram:

<img src="data/images/tab15.png" width="450">

In the image above, calculations will be performed at the category level rather than at the state level. However there may be cases where state level calculations may be desired along with the category calcs. The only way to do this without removing category altogether is to use **LOD calculations as they ARE CALCULATED INDEPENDENT OF THE VIEW ITSELF**

**2. Detail Button**
The second method for determining level of detail is the Detail shelf on the Mards card. This is commonly referred to as the Detail button. 

<img src="data/images/tab16.png" width="150">

**3. LOD Calculations**
The third method possible for determining level of detail. These calculations DO NOT affect the dimensions within a visualization, thus allowing one to set the level of detail desired indpendent of the view. 

There are 3 types of LOD expression in Tableau:
* **INCLUDE** - Calculating at a lower LOD
* **FIXED**  - Specifying the exact LOD **This is the default if none of the three expressions are chosen**
* **EXCLUDE** - Calculating at a higher LOD

## 7c. [Parameters](https://help.tableau.com/current/pro/desktop/en-us/parameters_create.htm)

[Parameters Tutorial](https://interworks.com/blog/rcurtis/2016/04/20/tableau-deep-dive-parameters-parameter-overview/)

A parameter is a workbook variable such as a number, date, or string that can replace a constant value in a calculation, filter, or reference line.

For example, you may create a calculated field that returns True if Sales is greater than $500,000 and otherwise returns False. You can replace the constant value of “500_000” in the formula with a parameter. Then, using the parameter control, you can dynamically change the threshold in your calculation.

**Bottom line is, paramaters are akin to variables in programming that can be adjusted in real-time by the user.**

You can even create a dynamic parameter that’s set to automatically refresh its current value (to the result of a single-value, view-independent calculation), list of values (based on a data source column), or range of values. This will happen each time the workbook is opened and Tableau connects to the data source referenced by the parameter, or whenever you select Refresh from the data source’s context menu.

You can make your parameters more dynamic and interactive by using them in **Parameter Actions**. Parameter actions let your audience change a parameter value through direct interaction with a viz, such as clicking or selecting a mark.

There are four standard use cases for parameters:

* Filters
* Bins
* Reference Lines
* Calculated Fields

## 10. Dashboards and Interactivity

[Good link for tutorials](https://interworks.com/blog/rcurtis/2017/05/22/tableau-deep-dive-dashboard-design-adding-interactivity/?_fsi=)


There are a lot of great things you can do within Tableau Desktop to add interactivity. You can even add combinations of these elements for really cool features. They break down into four basic varieties: 

**GLOBAL FILTERS:**  are standard filters that are created from one of the views in your desktop and then applied to more than the original worksheet. The term global filter implies that the filter affects every single view in that dashboard, but that’s not necessarily the case. They could apply to as little as two views, the original view and another.

**PARAMETERS:** Parameters are one of the most powerful elements in Tableau Desktop, because they can be used in a lot of different cases. Here are the main ones: calculated fields, histogram bins, reference lines, top N filters, and customized table calculations. With all of these options, parameters can do all sorts of things, including completely swapping out the measures and dimensions in your view. This is how you can add multiple views within a single visualization with just a single parameter control.

**DASHBOARD ACTIONS:** Dashboard actions are unique features for dashboards. They allow worksheets the ability to interact outside of just the domain of their view. There are three types of dashboard actions: filter actions, highlight actions, and URL actions. These are created and accessed throgh Dashboard ---> Actions ---> Add Action

* **Filter Actions -** allow the selections in one vizulazation to affect others. This can even be used to navigate between different dashboards

* **Highlight Actions -** allow you to emphasize the marks in other visualizations by interacting with the source worksheet The highlight will appear in the same visual style as the Highlighter tool or clicking on a value in the color legend

* **URL Actions -** help the user navigate to a web address from within the dashboard

**TOOLTIPS:** are details that appear when you rest the pointer over one or more marks in the view

## Really good explanation on how to create new column conditionally
[Create new column with conditional values](https://stackoverflow.com/questions/19913659/pandas-conditional-creation-of-a-series-dataframe-column)

```python
conditions = [(df['CreditScore'] < 580),
              (df['CreditScore'] >= 580) & (df['CreditScore'] < 689),
              (df['CreditScore'] >= 670) & (df['CreditScore'] < 749),
              (df['CreditScore'] > 749)]
choices = ['bad', 'fair', 'good', 'excellent']

df['Rating'] = np.select(conditions, choices, default='NOSCORE')

df.info()
```
